```python
import os
import numpy as np
import cv2 # OpenCV
from sklearn.svm import SVC # SVM klasifikator
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier # KNN
from joblib import dump, load
import matplotlib
import matplotlib.pyplot as plt


from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2


%matplotlib inline
# prikaz vecih slika 
matplotlib.rcParams['figure.figsize'] = 16,12
```

```python
TRAIN_DATASET_PATH = '.' + os.path.sep + 'dataset' + os.path.sep + 'train' + os.path.sep
VALIDATION_DATASET_PATH = '.' + os.path.sep + 'dataset' + os.path.sep + 'validation' + os.path.sep
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
image = cv2.imread(os.path.join(VALIDATION_DATASET_PATH, 'image_0.jpg'))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
rects = detector(gray, 1)
print(rects)
rect = (x, y, w, h) = face_utils.rect_to_bb(rects[0])
print(rect)
print(image.shape)
if y < 0:
    h = h + y
    y = 0
if x < 0:
    w = w + x
    x = 0
if x + w >= image.shape[1]:
    w = image.shape[1] - x - 1
if y + h >= image.shape[0]:
    h = image.shape[0] - y - 1
rect = (x, y, w, h)
print(rect)
cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)
plt.imshow(image)
```

```python
shape = predictor(gray, rects[0])
shape = face_utils.shape_to_np(shape)
for (x, y) in shape:
    cv2.circle(image, (x, y), 1, (0, 0, 255), -1)
cv2.circle(image, (x, y), 1, (0, 0, 255), -1)
plt.imshow(image)
```
