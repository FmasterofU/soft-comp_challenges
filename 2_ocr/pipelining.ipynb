```python
def histogram(image, xmax):
    height, width = image.shape[0:2]
    x = range(0, xmax+1)
    y = np.zeros(xmax+1)
    for i in range(0, height):
        for j in range(0, width):
            pixel = image[i, j]
            y[pixel] += 1
    return x, y
```

```python
VALIDATION_DATASET_PATH = '.'+os.path.sep+'dataset'+os.path.sep+'validation'+os.path.sep
num = 0
for image_path in glob.glob(VALIDATION_DATASET_PATH + "*.png"):
    cvimg = cv2.imread(image_path)
    hsvImage = cv2.cvtColor(cvimg, cv2.COLOR_BGR2HSV)
    x, y = histogram(hsvImage[:, :, 0], 179)
    ypick = y.copy()
    #pixels = hsvImage.shape[0] * hsvImage.shape[1]
    pixelsmax = np.max(ypick)
    for i in range(len(y)):
        if (i!=0 and ypick[i-1]>ypick[i]) or (i!=179 and ypick[i]<=ypick[i+1]):
            ypick[i] = 0
    for i in range(len(y)):
        if 0.05 * pixelsmax > ypick[i]:
            ypick[i] = 0
    #pickpeaks, _ = scipy.signal.find_peaks(ypick)
    pickpeaks = np.nonzero(ypick)[0]
    print(pickpeaks)
    if len(pickpeaks) == 2:
        print("TAKE:",y[pickpeaks[0]],y[pickpeaks[1]])
        #plt.figure()
        print(image_path)
        #plt.plot(x, y, '-or') 
        #plt.show()
        plt.figure()
        peak = min(pickpeaks.tolist(), key=lambda x: ypick[x])
        textimg = np.zeros(hsvImage[:, :, 0].shape, hsvImage[:, :, 0].dtype)
        textimg[hsvImage[:,:,0]==peak] = 255
        plt.imshow(textimg, 'gray')
        plt.show()
        num+=1
    else:
        plt.figure()
        print(image_path)
        print("NOT", len(pickpeaks))
        plt.plot(x, y, '-or')
        plt.show()
print(num)
```
